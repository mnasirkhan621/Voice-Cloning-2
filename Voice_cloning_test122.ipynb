```python
import uvicorn
import logging
import nest_asyncio
import sys
import torch
import torchaudio
import io
from fastapi import FastAPI, File, UploadFile, Form, HTTPException
from fastapi.responses import StreamingResponse

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(name)s - %(levelname)s - %(message)s",
    handlers=[logging.StreamHandler(sys.stdout)]
)
logger = logging.getLogger(__name__)

# Apply nest_asyncio for Jupyter compatibility
nest_asyncio.apply()

# FastAPI app
app = FastAPI(title="Voice Cloning API")

# Placeholder for model initialization (replace with actual model)
# model = YourVoiceCloningModel()  # Uncomment and initialize your model here
# For testing, assume a dummy model with sample rate
class DummyModel:
    sr = 44100  # Replace with actual model sample rate
model = ChatterboxTTS.from_pretrained(device=device)  #

@app.post("/clone-voice/")
async def clone_voice(
    text: str = Form(...),
    reference_audio: UploadFile = File(...)
):
    try:
        # Input validation
        if not text.strip():
            raise HTTPException(status_code=400, detail="Text input cannot be empty")
        if not reference_audio.filename.lower().endswith((".wav", ".mp3")):
            raise HTTPException(status_code=400, detail="Reference audio must be WAV or MP3")

        # Load and preprocess reference audio
        logger.info("Loading reference audio...")
        audio_bytes = await reference_audio.read()
        audio_buffer = io.BytesIO(audio_bytes)
        waveform, sample_rate = torchaudio.load(audio_buffer)
        logger.info(f"Reference audio - Shape: {waveform.shape}, Sample rate: {sample_rate}, Dtype: {waveform.dtype}")

        # Resample audio if necessary
        if sample_rate != model.sr:
            logger.info(f"Resampling audio from {sample_rate} Hz to {model.sr} Hz")
            resampler = torchaudio.transforms.Resample(sample_rate, model.sr)
            waveform = resampler(waveform)

        # Convert to mono if stereo
        if waveform.shape[0] > 1:
            logger.info(f"Converting to mono (original channels: {waveform.shape[0]})")
            waveform = waveform.mean(dim=0, keepdim=True)

        # Normalize audio amplitude
        if waveform.abs().max() > 0:
            waveform = waveform / waveform.abs().max() * 0.9
            logger.info("Normalized reference audio amplitude")

        # Save preprocessed audio
        audio_buffer = io.BytesIO()
        torchaudio.save(audio_buffer, waveform, model.sr, format="wav")
        audio_buffer.seek(0)

        # Generate audio
        logger.info(f"Generating audio for text: {text[:50]}...")
        wav = model.generate(text, audio_prompt_path=audio_buffer)  # Replace with actual model call
        logger.info(f"Generated waveform - Shape: {wav.shape}, Dtype: {wav.dtype}, Device: {wav.device}")

        # Handle tensor dimensions
        if wav.ndim == 3:
            logger.info(f"3D tensor detected, reducing to 2D. Shape: {wav.shape}")
            # Try different reduction strategies
            if wav.shape[0] == 1:
                wav = wav[0]  # [1, channels, samples] -> [channels, samples]
            elif wav.shape[1] == 1:
                wav = wav[:, 0, :]  # [batch, 1, samples] -> [batch, samples]
            elif wav.shape[2] == 1:
                wav = wav[:, :, 0]  # Rare case: [batch, channels, 1]
            else:
                raise ValueError(f"Cannot reduce 3D tensor to 2D: shape {wav.shape}")
        elif wav.ndim == 1:
            logger.info(f"1D tensor detected, converting to 2D. Shape: {wav.shape}")
            wav = wav.unsqueeze(0)  # [samples] -> [1, samples]
        elif wav.ndim != 2:
            raise ValueError(f"Expected 2D tensor, got {wav.ndim}D: shape {wav.shape}")

        logger.info(f"Processed waveform - Shape: {wav.shape}, Dtype: {wav.dtype}")

        # Ensure float32 and CPU
        if wav.dtype != torch.float32:
            logger.info(f"Converting to float32 from {wav.dtype}")
            wav = wav.to(torch.float32)
        wav = wav.cpu()

        # Save to output buffer
        output_buffer = io.BytesIO()
        torchaudio.save(output_buffer, wav, model.sr, format="wav")
        output_buffer.seek(0)

        logger.info("Audio generated successfully")
        return StreamingResponse(
            output_buffer,
            media_type="audio/wav",
            headers={"Content-Disposition": "attachment; filename=output.wav"}
        )

    except Exception as e:
        logger.error(f"Endpoint error: {str(e)}")
        raise HTTPException(status_code=500, detail=f"An internal error occurred: {str(e)}")

def run_server():
    try:
        logger.info("Starting Uvicorn server on http://0.0.0.0:8015")
        uvicorn.run(
            app,
            host="0.0.0.0",
            port=8015,
            log_level="info",
            timeout_keep_alive=30,
            access_log=True,
            loop="asyncio"  # Explicitly use asyncio loop
        )
    except Exception as e:
        logger.error(f"Failed to start Uvicorn server: {str(e)}")
        raise

if __name__ == "__main__":
    print("Starting FastAPI server...")
    print("API will be running at http://localhost:8015")
    print("Interactive documentation at http://localhost:8015/docs")
    print("\nTo stop, interrupt the cell (Ctrl+C) or restart the Jupyter kernel.")
    try:
        run_server()
    except KeyboardInterrupt:
        logger.info("Server stopped by user")
        print("Server stopped gracefully.")
    except Exception as e:
        logger.error(f"Server error: {str(e)}")
        print(f"Failed to start server: {str(e)}")
```
